```python
# read the HDF result 
# 2021/04/27
import h5py
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from datetime import datetime
```

```python
file_path = "E:\Black_earth_hec\Hec_2d\BEC"
file_name = "Black_Earth_Creek_0.p04.hdf"
hf = h5py.File(file_path + "\\" + file_name, "r")
```

```python
def stage_extraction(hf, cell_number):
    # read time stamp
    time_stamp = hf["Results/Unsteady/Output/Output Blocks/Base Output/Unsteady Time Series/Time Date Stamp"]
    # read water surface
    water_surface = hf["Results/Unsteady/Output/Output Blocks/Base Output/Unsteady Time Series/2D Flow Areas/BEC/Water Surface"]
    # read node stage time series
    node_surface = water_surface[:, cell_number]
    return time_stamp, node_surface
```

```python
cell_number = 30976
node_surface = stage_extraction(hf, cell_number)
plt.plot(time_stamp, node_surface)
```

```python
# 2021/05/07 Read HDF output file plan 16
# verify with gage measurement stage data
simulation = pd.DataFrame() # store simulation data
```

```python
# load simulation results
file_path = "E:\Black_earth_hec\Hec_2d\BEC_0428"
file_name = "BEC_0428.p16.hdf"
hf = h5py.File(file_path + "\\" + file_name, "r")
# load processed treatmentplant gage data
file_location = r"D:\Manning_gaussian\Black_earth_creek\Unsteady_flow_data\Stage\treatment_plant_stage_2018_08_19_2018_08_26_15min.csv"
treatment_stage = pd.read_csv(file_location)
```

```python
# read cells at gage treatment plant
treatment_plant_node = 45294 # 45302
time_stamp, node_surface = stage_extraction(hf, treatment_plant_node)
# change from byte to utf-8
index_dss = []
for i in range(len(time_stamp)):
    time_stamp_str = time_stamp[i].decode('UTF-8')
    d = datetime.strptime(time_stamp_str, '%d%b%Y %H:%M:%S')
    index_dss.append(d)
simulation[treatment_plant_node] = node_surface
simulation.index = index_dss
# read cell stage at black earth
black_earth_node = 30870
time_stamp, node_surface = stage_extraction(hf, black_earth_node)
simulation[black_earth_node] = node_surface # store stage
```

```python
index = []
for i in range(treatment_stage.shape[0]):
    d = datetime.strptime(treatment_stage['Unnamed: 0'][i], '%Y-%m-%d %H:%M:%S') # convert string to datetime object
    index.append(d)
treatment_stage.index = index
```

```python
plt.figure(figsize = (20, 10))
plt.plot(simulation[treatment_plant_node], label = 'HECRAS')
plt.plot(treatment_stage["Stage(feet)"], label = 'Observation')
plt.plot(simulation[treatment_plant_node][10:30], 'o')
plt.plot(treatment_stage["Stage(feet)"][50:120],'o')
warm_ras_stage = simulation[treatment_plant_node][10:30].mean() # still state data
warm_obs_stage = treatment_stage["Stage(feet)"][50:120].mean() # still state data
plt.plot(simulation[treatment_plant_node] + warm_obs_stage - warm_ras_stage, label = 'HECRAS_offset')
plt.legend()
# plt.plot(index, treatment_stage["Stage(feet)"])
```
